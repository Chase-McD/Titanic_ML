# Titanic Decision Tree Classifier

This algorithm tests predicts if a passsenger would survive or pass away based on the variables [Pclass,sex,age,sibsp,parch,fare,embarked].

Note: in this case 0 = died and 1 = survived. This will be important to know when adding inputs [Pclass,sex,age,sibsp,parch,fare,embarked] into the model to predict if someone died or passed away.

```python
import pandas as pd
df = pd.read_csv('tested.csv')
df
```

```python
df = df.drop(columns = ['Cabin'])
df = df.dropna()
df
```

```python
#make the strings dummies
df = pd.get_dummies(df, columns=["Sex","Embarked"])
df

#check to see if there are any remaining missing fields of information
df.isnull().sum().sum()
```

```python
X = df.drop(columns = ["PassengerId","Survived","Name","Ticket"])
y = df["Survived"]
X
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X,y)
```

```python
#Make a prediction
predictions = model.predict([[3,34.5,0,0,7.8292,0,1,0,1,0]])
print(predictions)

#This person was predicted by the model to die. looking at the data we can see that a person with these same credintials did truly die as in the first person (index = 0) in the df
print(y.head())
X.head()
```

```python
#Testing the model

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print(y_test)
print(predictions)


score = accuracy_score(y_test, predictions)
score
```
